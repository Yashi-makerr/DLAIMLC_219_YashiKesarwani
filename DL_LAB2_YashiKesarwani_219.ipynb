```python
X=[0.45,0.38,0.60,0.23,0.76]
w=[0.4,0.2,0.3,0.1,0.7]
b=-0.2
Y=1
n=0.01
```

```python
z=0
for i in range(len(X)):
    z += w[i]*X[i]

z += b
```

```python
if z>=0:
    y_pred=1
else:
    y_pred=0
```

```python
error=Y-y_pred
```

```python
print(error)
```

```python
for i in range(len(w)):
    w[i]=w[i]+n*error*X[i]
```

```python
b=b+n*error
```

```python
print("Net input:",z)
print("Predicted:",y_pred)
print("Error:",error)
print("Updated weights:",w)
print("Updated bias:",b)
```
